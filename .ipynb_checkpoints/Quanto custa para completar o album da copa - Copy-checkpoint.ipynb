# Estudo: Quanto custa para completar o álbum da copa? 
## Uma abordagem probabilística prática.

O álbum da copa de 2022 começou a ser vendido no Brasil em 19 de agosto de 2022, e conta com 670 figurinhas, que são vendidas em pacotes de 5 por R$ 4,00.

Com essas informações, conseguimos calcular quantos pacotes de figurinha teríamos que comprar:

$$ Número\ necessário\ de\ pacotes = \frac{670}{5} = 134 $$

Sabendo quantos pacotes temos que comprar, podemos calcular o quanto precisaríamos gastar: <br>

$$134 \times 4 = 536$$

Ou seja, para comprar todas as figurinhas, precisaríamos gastar R$ 536,00... O problema é que, como todos sabem, teremos figurinhas repetidas no meio dessas que comprarmos, então, a chance de você conseguir obter 670 figurinhas não repetidas é extremamente baixa. Inclusive, podemos quantificar essa chance facilmente com um pouquinho de matemática.

## Usando um pouquinho de probabilidade...

Aqui, vamos calcular rapidamente, qual é a chance de obtermos 670 figurinhas não repetidas. Suponhamos que uma pessoa comprou seu álbum, e por ser ansiosa, já comprou os 134 pacotes esperando completar o álbum de uma só vez. Então ela chega em casa, abre todos os pacotes rapidamente, empilha todas as figurinhas e vai começar a colá-las no álbum. Qual é a chance de que, retirando as figurinhas 1 a 1 da pilha e colando elas no álbum, esta pessoa não encontre nenhuma figurinha repetida?

Vamos usar um poquinho de probabilidade para calcular isso. E começo a fazer isso lhe fazendo, leitor, a seguinte pergunta: Qual é a chance de que a primeira figurinha **não** seja repetida (leia-se: não esteja colada no álbum)?

Bom, acredito que seja óbvio que essa resposta seja **100%**! Ou seja, a chance de que a primeira figurinha que vou colar no meu álbum não seja repetida é de 100%.

Esses 100% vem da seguinte conta:

$$ Número\ de\ figurinhas\ que\ não\ tenho = 670 $$ <br>
$$ Espaço\ amostral = 670 $$ <br>
$$ 670 / 670 = 1 = 100\% $$

A conta acima pode ser lida da seguinte forma: Não tenho nenhuma figurinha colada no meu álbum ainda (faltam 670), porque vou começar a completar meu álbum agora, e eu tenho um total de 670 figurinhas diferentes que podem ser retiradas da pilha a cada momento. Para eu ter a probabilidade deste evento, basta eu dividir o $Número\ de\ figurinhas\ não\ repetidas$ pelo tamanho do meu $Espaço\ amostral$ (que é o número de figurinhas existentes), isso me retorna o valor de 1, que é igual a 100%!

Beleza, a pessoa retirou a primeira figurinha, e como calculamos, ela não era repetida... Agora, qual é a chance da segunda figurinha não ser repetida? Ora, podemos aplicar a mesma lógica acima e calcular:

$$ Número\ de\ figurinhas\ que\ não\ tenho = 669 $$ <br>
$$ Espaço\ amostral = 670 $$ <br>
$$ 669 / 670 = 0.9985 = 99,85\% $$

Agora vemos que a chance de a próxima figurinha ser uma que não está colada no álbum é de 99,85%. Ou seja, seria uma baita falta de sorte já ter uma figurinha repetida ao tirar a segunda figurinha da pilha.

A questão é que, por ora, por mais que tenhamos a probabilidade "em nosso favor" porque é alta a chance de estarmos tirando figurinhas novas (não repetidas) da pilha, conforme formos completando o nosso álbum, as chances de termos figurinhas não repetidas vai diminuindo, e gradativamente deixando de estar "em nosso favor".

Agora indo diretamente à questão, qual a chance de as 670 figurinhas da pilha serem diferentes umas das outras? Podemos calcular isso de forma relativamente simples, por meio da probabilidade de eventos sucessivos:

$$ \frac{670}{670} \times \frac{669}{670} \times \frac{668}{670} \times\ ...\ \times \frac{3}{670} \times \frac{2}{670} \times \frac{1}{670}$$

Entenda a conta acima da seguinte forma: cada fração mostrada é a probabilidade de retirarmos uma figurinha não repetida, até chegarmos na última figurinha, e eu estou multiplicando essas frações porque estou tentando calcular a probabilidade de eventos sucessivos, ou seja, de tirarmos uma figurinha não repetida, depois outra, depois outra, depois outra...

Fazer essa conta manualmente é bem complicado... Mas fazer isso em Python é bem simples:

```python
from functools import reduce
quantidade_total_de_figurinhas = 670

# Aqui eu tenho a sequência das probabilidades:
lista_de_probabilidades = [(670-i)/670 for i in range(quantidade_total_de_figurinhas)] 

# E eu multiplico todas elas:
probabilidade = reduce(lambda x, y : x*y, lista_de_probabilidades)
```

```python
probabilidade
```

Ok, acima podemos ver que a probabilidade de isso acontecer é um número **extremamente** baixo, leia-se ~$6.8*10^{-300}$, ou seja, não conte com isso acontecendo...

## O problema do colecionador de cupons

Aproveitando que estamos falando de probabilidade, vou tocar em um assunto que surge com certa frequência quando se pesquisa sobre álbuns de figurinhas e os custos e probabilidades envolvidas. E este assunto é o [Problema do Colecionador de Cupons](https://pt.wikipedia.org/wiki/Problema_do_colecionador_de_cupons)!

De forma resumida, porque quero propor uma abordagem diferente, o **Problema do Colecionador de Cupons** nos traz uma pergunta que tem como resposta algo muito semelhante com o que queremos calcular aqui. Sem me estender muito, ao estudar o problema do colecionador de cupons, nós podemos calcular o valor esperado da quantidade de pacotes de figurinhas que teríamos que comprar para completar o álbum.

E isso pode ser calculado ao inverter as probabilidades (frações) que eu havia calculado lá em cima, e somando-as, ou seja:

$$ \frac{670}{670} + \frac{670}{669} + \frac{670}{668} + ... + \frac{670}{3} + \frac{670}{2} + \frac{670}{1}$$

Para fazer isso em Python, é bem simples e só basta eu modificar um pouco o código que eu havia feito acima:

```python
lista_de_chances = [670/(670-i) for i in range(quantidade_total_de_figurinhas)]
```

```python
sum(lista_de_chances)
```

Porém, antes de eu me aprofundar neste resultado, preciso explicitar algumas limitações do **Problema do Colecionador de Cupons**... Ele não contabiliza o fato de que as figurinhas vêm em pacotes de 5, então, o número acima tem o seguinte significado: *Precisaríamos comprar aproximadamente 4747 figurinhas até conseguir comprar as 670 diferentes figurinhas para completar o álbum*.

Já que cada pacote de figurinhas custa 4 reais e contém 5 figurinhas, podemos assumir que o custo de cada figurinha é de 80 centavos e, com uma simples matemática: 
$$ 0,80 \times 4747 = 3797,60 $$

Ou seja, com algumas simplificações, podemos usar o **Problema do Colecionador de Cupons** para estimar o valor esperado do custo para completar o álbum da copa de 2022, sem considerar, é claro, as trocas de figurinha e etc...

Porém, sabemos que é muito improvável que alguém vá completar o álbum sem realizar nenhuma troca, ou talvez sem pedir nenhuma figurinha para a Panini quando estiver finalizando e etc. Por isso, vou propor uma abordagem bem mais prática para o problema...

## Abordando o problema de uma forma prática...

O que vamos fazer agora é usar a linguagem Python e o poder computacional para abordar esse problema de uma forma muito mais prática. Eu vou escrevendo o código e explicando ao longo do caminho. Não se preocupe se você não estiver entendendo muito bem a parte da programação, prometo que vou tentar ser bem detalhista e focar mais nos resultados.

Primeiro eu vou realizar a importação de algumas bibliotecas que vão me ajudar na geração do código:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Antes de eu iniciar o código em si, acho que vale a pena eu explicar o que vou tentar fazer... Basicamente eu vou tentar simular a compra de pacotinhos de figurinhas, e verificar quantos pacotes eu tenho que comprar até eu conseguir comprar todas as diferentes figurinhas. E podemos ir rodando esse cenário para diferentes quantidades de figurinhas, como se fossêmos trocando algumas figurinhas ao longo do tempo.

Existem diferentes maneiras de simular essas compras, poderíamos criar ferramentas interessantes com a biblioteca `random` e usar o `Counter`, por exemplo, porém eu vou usar a própria biblioteca `Pandas` e o método `sample` dos `DataFrames`.

Primeiramente, preciso criar uma lista com todas as figurinhas, para que depois eu possa retirar amostras dessa lista, simulando a compra de um pacote de figurinhas:

```python
df_figurinhas = pd.DataFrame({'figurinha':range(1,671)})
df_figurinhas
# Veja a lista de figurinhas:
```

```python
def compra_pacote(df):
    '''
    Aqui eu estou criando uma função que simula a compra de um pacote de figurinhas.
    Usando o método sample do DataFrame.
    O único argumento da função é o dataframe utilizado.
    
    '''
    
    pacote = np.array(df.sample(n=5, replace=False, weights=None)['figurinha']) # 5 figurinhas, não podem ser repetidas no mesmo pacote, todas com igual chance de estarem no pacote
    
    return pacote
```

Acima eu criei uma função que permite a simulação da compra de um pacote de figurinhas, veja abaixo:

```python
list(compra_pacote(df_figurinhas))
```

Simulamos a compra de um pacote que contém as respectivas figurinhas apresentadas acima! E para isso temos algumas premissas:

- Todas as figurinhas tem igual chance de aparecer nos pacotes.
- Não é possível obter duas figurinhas iguais no mesmo pacote.

Agora, o que podemos fazer é criar um código que vai "comprar pacotes de figurinhas" em loop, até conseguir completar o álbum:

```python
def completa_album(figurinhas_trocadas=0):
    
    '''
    Função para simular a compra de pacotes de figurinhas até que completemos o álbum.
    Esta função pode receber o número esperado de figurinhas trocadas, que serão subtraídas do total de figurinhas necessárias a serem compradas
    O retorno da função é a quantidade de pacotes de figurinhas que foram comprados para completar o álbum
    '''

    i=0 # Contador que uso para contar a quantidade de pacotes "comprados"
    arr_fig_obtidas = np.empty((1,0)) # Arr que eu vou usar para armazenar as figurinhas que eu obtive

    while True:
        
        arr_curr = compra_pacote(df_figurinhas) # Aqui eu simulo a compra de um pacote de figurinhas
        i+=1 # Assim que eu "compro" o pacotinho, eu aumento meu contador em 1
        arr_fig_obtidas = np.append(arr_fig_obtidas, arr_curr) # Aqui eu adiciono as figurinhas desse pacote a uma lista de figurinhas obtidas
        figurinhas_faltantes = 670 - len(np.unique(arr_fig_obtidas)) - figurinhas_trocadas # Aqui eu vejo quantas figurinhas faltam
        #print(f'{i} pacotinhos comprados, faltam {figurinhas_faltantes}')
        if figurinhas_faltantes <= 0: # Caso não falte nenhuma:
            break # Paro de comprar pacotinhos

    return i # E eu retorno a quantidade de pacotinhos que compramos!
```

Acima eu criei uma função que vai simular a compra de pacotes de figurinhas, até que tenhamos as 670 figurinhas necessárias para completar o álbum. E, eu adicionei um argumento opcional na função, onde podemos inserir a quantidade de figurinhas trocadas, para termos diferentes cenários, e conseguir observar como o custo vai diminuindo conforme vamos trocando figurinhas.

Posso rodar a função uma vez para exemplificar:

```python
completa_album()
```

Acima, temos a quantidade de pacotes que foram comprados para completar o álbum, mas, por ora, peço que não se apegue ao número mostrado, porque estamos falando de apenas um álbum, e como estamos lidando com um cenário envolvendo probabilidades e um certo nível de aleatoriedade, esse número vai mudar cada vez que rodarmos a função. Então, para que tenhamos um número mais confiável, o que eu posso fazer é rodar a função uma determinada quantidade de vezes e verificarmos qual é a quantidade média de pacotes.

Isso é bem fácil de fazer:

```python
def completa_n_albuns(n=1, figurinhas_trocadas=0):
    '''
    Aqui crio uma função que vai completar n álbuns e retornar uma lista com a quantidade de pacotes necessários para completar cada álbum!
    '''


    lista_pacotes = [completa_album(figurinhas_trocadas) for i in range(n)]

    return lista_pacotes
```

Agora, temos praticamente tudo que precisamos para fazer a simulação completa! Vamos criar um cenário de 300 álbuns sendo completos, ainda nos limitando a não trocar nenhuma figurinha:

```python
lista_pacotes_necessarios = completa_n_albuns(n=300)
```

Agora que relizamos a simulação, completando 300 álbuns sem realizar troca, podemos ver a distribuição dos pacotinhos para completar os álbuns:

```python
fig, ax = plt.subplots(figsize=(12,6), dpi=500)

sns.histplot(
    data=lista_pacotes_necessarios,
    ax=ax,
    bins=50
)

plt.title('Distribuição dos pacotes da quantidade de pacotes de figurinhas para completar o álbum da copa de 2022 sem trocas')
plt.ylabel('Número de ocorrências')
plt.xlabel('Quantidade de Pacotes')
```

Acima, podemos ver que alguns álbuns precisaram de mais pacotes do que outros, etc, como esperado, e o valor médio da distribuição, que é o que nos interessa, é:

```python
quantidade_media_pacotes = round((sum(lista_pacotes_necessarios) / len(lista_pacotes_necessarios)), 0)
quantidade_media_pacotes
```

Ou seja, para completar o álbum, sem trocar nenhuma figurinha, esperaríamos gastar:

```python
quantidade_media_pacotes * 4
```

Que comprova, aproximadamente, o valor trazido pelo **Problema do Colecionador de Cupons**!!

Como estamos falando de probabilidades, e, conforme você pode ver na distribuição acima, existem álbuns que precisaram de mais figurinhas, e outros que precisaram de menos figurinhas. O que podemos fazer também, é verificar os piores casos, que seriam as barrinhas presentes no extremo direito da distribuição. Vamos verificar os 5% ábuns mais caros:

```python
# Aqui eu vou filtrar o meu array para trazer apenas os valores acima do percentil de 95%

# Calculo qual é o percentil:
percentil_95 = np.quantile(lista_pacotes_necessarios, 0.95)

# Crio uma lista auxiliar apenas com os valores acima do percentil de 95%:
lista_pacotes_necessarios_percentil_95 = np.array(lista_pacotes_necessarios)[lista_pacotes_necessarios > percentil_95]

# Retorno a média do percentil, multiplicado pelo preço do pacote de figurinhas:
np.mean(lista_pacotes_necessarios_percentil_95) * 4
```

Acima, você pode ver o valor médio de 5% dos álbuns mais caros a se completar, quando não realizarmos a troca de nenhuma figurinha

Porém, vamos finalmente ao que interessa: **como que esse número diminui conforme vamos trocando um certo número de figurinhas?**

O que podemos fazer para simular esses cenários, é diminuir artificialmente a quantidade de figurinhas necessárias para completar o álbum, ou seja, caso você troque 50 figurinhas, vou considerar que você precisa de 620 (670 - 50) para completar seu álbum, e assim por diante. E podemos então, rodar a mesma simulação acima, para esses diferentes cenários, e ver como o valor vai diminuindo.

## Rodando diferentes cenários:

```python
lista_cenarios = list(range(0, 410, 10))
```

Acima, eu criei uma lista com diferentes cenários, desde trocando 0 figurinhas, até trocando 400 figurinhas. E eu vou aumentando a quantidade de 10 em 10. Isso resulta em 41 cenários diferentes.

Abaixo, eu consigo criar um mecanismo de *loop* que vai "completar" 100 álbuns para cada um desses cenários, e me retornar a quantidade média de pacotes de figurinhas para cada um dos cenários:

```python
df_cenarios = pd.DataFrame()

for cenario in lista_cenarios:
    # Vou iterar por todos os cenários criados

    lista_pacotes_necessarios = completa_n_albuns(n=100, figurinhas_trocadas=cenario) # "Completo" 100 álbuns para cada cenário 

    qtd_pacotes_necessarios = int(round((sum(lista_pacotes_necessarios) / len(lista_pacotes_necessarios)), 0)) # Armazeno a quantidade média de pacotes necessários
    cenario_atual = {'cenario':cenario, 'qtd_pacotes_necessarios':qtd_pacotes_necessarios}

    df_cenarios = df_cenarios.append(cenario_atual, ignore_index=True) # Coloco o cenário atual no DataFrame

    print(f'Cenário com {cenario} figurinhas trocadas gerado...') # Printo uma mensagem para ir acompanhando o andamento do processo
```

Tendo rodado os cenários, eu tenho a seguinte tabela, que armazenou a quantidade de pacotes necessários para cada um deles:

```python
df_cenarios
```

Observe como a queda na quantidade de pacotes necessários é brusca logo no início da tabela. Um gráfico pode ajudar na visualização:

```python
fig, ax = plt.subplots(figsize=(12,6), dpi=400)

sns.lineplot(
    data=df_cenarios,
    x='cenario',
    y='qtd_pacotes_necessarios',
    ax=ax
)

plt.title('Quantidade de pacotes necessários para completar o álbum em função da quantidade de figurinhas trocadas')
plt.xlabel('Qtd. figurinhas trocadas')
plt.ylabel('Qtd. média de pacotes comprados')
```

O problema é que, por mais que a modelagem aponte que precisaríamos de mais ou menos 70 pacotes de figurinhas quando trocamos 400 figurinhas, por exemplo. Eu comprar apenas 70 pacotes não me dará a quantidade de figurinhas o suficiente para trocar e conseguir completar o álbum! Portanto, por mais que a nossa simulação aponte cenários onde eu vou precisar **comprar menos do que 134 pacotes**, isso não necessariamente se aplicaria à realidade, dado que, se eu comprar menos que isso, **eu não vou ter figurinhas o suficiente para colar no álbum e depois trocar o restante!**

Para entender isso melhor, vamos pegar de exemplo, o cenário onde eu realizaria a troca de 250 figurinhas. Neste cenário, precisaríamos encontrar 420 (670 - 250) figurinhas não repetidas por meio da compra de pacotes, e a modelagem apontou que para encontrar as 420 figurinhas não repetidas, precisamos comprar 133 pacotes, que nos dá um total de 665 figurinhas (133 * 5). Porém, temos um problema, pois, isso é menos do que o suficiente para conseguirmos de fato trocar as 250, pois se tivermos 420 (não repetidas) coladas no álbum, e comprado apenas 133 pacotes, nos sobrariam 245 figurinhas para trocar, e se eu trocar todas elas, ainda vão faltar 5 figurinhas para eu conseguir completar o álbum. E esse raciocínio é válido para todos os cenários posteriores.

Portanto, minha conclusão é que, por mais que você troque muitas figurinhas, você sempre vai precisar comprar pelo menos a quantidade de figurinhas necessárias para completar o álbum (670), porque você precisa, primeiramente colar as que não forem repetidas, e depois, pelo menos a quantidade necessária para trocar o restante 1 a 1 e conseguir o suficiente de figurinhas diferentes para completar o álbum.

Aplicando esta lógica, conseguimos fazer um ajuste na nossa tabela:

```python
df_cenarios['qtd_pacotes_necessarios_ajustado'] = df_cenarios.apply(lambda x : 134 if x['qtd_pacotes_necessarios'] < 134 else x['qtd_pacotes_necessarios'], axis=1)
```

Acima eu criei uma coluna nova na tabela, ajustando os cenários onde foi necessário comprar menos do que 134 pacotes, informando que em todos eles, por mais que em tese eu precisassse de menos, eu ainda precisaria dos 134 para ter figurinhas o suficiente para realizar as trocas:

```python
# Podemos plottar o gráfico novamente:

fig, ax = plt.subplots(figsize=(12,6), dpi=400)

sns.lineplot(
    data=df_cenarios,
    x='cenario',
    y='qtd_pacotes_necessarios_ajustado',
    ax=ax
)

plt.title('Quantidade de pacotes necessários (ajustado) para completar o álbum em função da quantidade de figurinhas trocadas')
plt.xlabel('Qtd. figurinhas trocadas')
plt.ylabel('Qtd. média de pacotes comprados')
```

Agora veja que, por mais que eu troque uma maior quantidade de figurinhas, o número de pacotes necessários não diminui, garantindo que eu sempre vou ter figurinhas o suficiente para realizar as trocas!

Finalmente, conseguimos calcular o custo esperado para cada um dos cenários:

```python
df_cenarios['custo'] = df_cenarios['qtd_pacotes_necessarios_ajustado'] * 4
```

```python
df_cenarios
```

Incluindo a coluna de custo, vemos que por mais que aumentemos a nossa quantidade de figurinhas trocadas, **o menor custo, em um cenário ideal, para completar o álbum é R$536,00**, para que tenhamos a quantidade de figurinhas o suficiente para colar e depois trocar o restante! E, nos piores casos, sem troca nenhuma de figurinha, nosso custo seria de quase 4 mil reais! 

Portanto, caso esteja colecionando o álbum, se prepare para gastar de acordo com o que está disposto a trocar de figurinhas, e troque o máximo possível!

Por fim, posso fazer um plot do custo esperado em cada cenário:

```python
# Podemos plottar o gráfico novamente:

fig, ax = plt.subplots(figsize=(15,6), dpi=400)

sns.barplot(
    data=df_cenarios,
    x=df_cenarios['cenario'].astype(int),
    y='custo',
    ax=ax
)

plt.tight_layout()
plt.title('Custo esperado (R$) para completar o álbum em função da quantidade de figurinhas trocadas')
plt.xlabel('Qtd. figurinhas trocadas')
plt.ylabel('Custo esperado em R$')
plt.xticks(rotation=-25)

def add_value_labels(ax, spacing=5):
    # Função adaptada do Stak Oveflow no link: https://stackoverflow.com/questions/28931224/how-to-add-value-labels-on-a-bar-chart
    # Esta função adiciona os value labels acima da barra no gráfico abaixo

    """Add labels to the end of each bar in a bar chart.

    Arguments:
        ax (matplotlib.axes.Axes): The matplotlib object containing the axes
            of the plot to annotate.
        spacing (int): The distance between the labels and the bars.
    """

    # For each bar: Place a label
    for rect in ax.patches:
        # Get X and Y placement of label from rect.
        y_value = rect.get_height()
        x_value = rect.get_x() + rect.get_width() / 2

        # Number of points between bar and label. Change to your liking.
        space = spacing
        # Vertical alignment for positive values
        va = 'bottom'

        # If value of bar is negative: Place label below bar
        if y_value < 0:
            # Invert space to place label below
            space *= -1
            # Vertically align label at top
            va = 'top'

        # Use Y value as label and format number with one decimal place
        label = "{:.0f}".format(y_value)

        # Create annotation
        ax.annotate(
            label,                      # Use `label` as label
            (x_value, y_value),         # Place label at end of the bar
            xytext=(0, space),          # Vertically shift label by `space`
            textcoords="offset points", # Interpret `xytext` as offset in points
            ha='center',                # Horizontally center label
            va=va)                      # Vertically align label differently for
                                        # positive and negative values.

add_value_labels(ax)
plt.savefig('Custo esperado para completar o album.png', bbox_inches='tight')
```
